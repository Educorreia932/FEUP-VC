# 13. Segmentation - thresholding

```python
import cv2
import matplotlib.pyplot as plt

def show_image(image):
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    plt.show()

plt.rcParams['figure.figsize'] = [8, 10]
```

### Compare the global thresholding â€“ threshold() - and adaptive thresholding - adaptiveThreshold() - for an image with varying illumination 

```python
image = cv2.imread("Images/mutilator.jpg", cv2.IMREAD_GRAYSCALE) 

show_image(image)
```

```python
# Global thresholding
_, threshold_1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

show_image(threshold_1)
```

```python
# Otsu's thresholding (after Gaussian filtering)
blur = cv2.GaussianBlur(image, (5, 5), 0)
_, threshold_2 = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

show_image(threshold_2)
```

```python
threshold_3 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

show_image(threshold_3)
```
