# 16. Segmentation - Meanshift

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def show_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

plt.rcParams['figure.figsize'] = [8, 10]
```

### Segment some color images using the meanshift algorithm â€“ `cv.pyrMeanShiftFiltering()`

```python
image = cv2.imread("Images/floating.jpg") 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

show_image(image)
```

```python
spatial_radius = 20
segmented_image = cv2.pyrMeanShiftFiltering(image, spatial_radius, 10)

show_image(segmented_image)
```

```python
color_radius = 50
segmented_image = cv2.pyrMeanShiftFiltering(image, spatial_radius, color_radius)

show_image(segmented_image)
```

```python
maximum_pyramid_level = 6
segmented_image = cv2.pyrMeanShiftFiltering(image, spatial_radius, color_radius, maximum_pyramid_level)

show_image(segmented_image)
```
