# 4. Image enhancement â€“ filtering

```python
import cv2
import os
import numpy as np

from matplotlib import pyplot as plt
from scipy import ndimage
```

### Take a noisy image and filter it, using:

### **a)** a mean filter

```python
# Opening a noisy image
image = cv2.imread("Images/lizard_noisy.bmp")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # Convert from BGR to RGB because we are using matplotlib to display the image

plt.imshow(image)
plt.title('Original (Noisy) Image')
plt.axis('off')
plt.show()
```

```python
mean = cv2.blur(image, (4,4))

plt.imshow(mean)
plt.title('Mean Filter')
plt.axis('off')
plt.show()
```

### **b)** a Gaussian filter

```python
gaussian = cv2.GaussianBlur(image, (5, 5), 0)

plt.imshow(gaussian)
plt.title('Gaussian Filter')
plt.axis('off')
plt.show()
```

### **c)** a median filter

```python
median = cv2.medianBlur(image, 5)

plt.imshow(median)
plt.title('Median Filter')
plt.axis('off')
plt.show()
```

### **d)** a bilateral filter

```python
bilateral = cv2.bilateralFilter(image, 15, 75, 75)

plt.imshow(bilateral)
plt.title('Bilateral Filter')
plt.axis('off')
plt.show()
```

**e)** a filter defined by you

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 

kernel_3x3 = (1 / 16 ) * np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]])

print(kernel_3x3)

convolution = ndimage.convolve(gray, kernel_3x3)

plt.imshow(convolution, cmap="gray")
plt.title("My 3x3 convolution w/Gaussian mask")
plt.axis('off')
plt.show()
```

If the $(1 / 14)$ factor is removed, the values won't be in the $0-255$ range.
