```python
import cv2
import numpy as np
import os

from matplotlib import pyplot as plt

def show_image(image, title="", cmap=""):
    plt.figure()
    plt.title(title)
    plt.imshow(image, cmap="gray", vmin=0)
    plt.axis('off')
    plt.show()
```

### **a)** Detect the edges of an image using the `Canny()` OpenCV function. **Suggestion:** use trackbars to select different low and high threshold for the hysteresis procedure and different aperture size for the `Sobel()` function.

```python
image  = cv2.imread("Images/lizard.bmp", 0)

show_image(image, title="Original image")
```

```python
threshold1 = 100
threshold2 = 200
apertureSize = 3

edges = cv2.Canny(image, threshold1, threshold2)

show_image(edges, title="Edge image")
```
